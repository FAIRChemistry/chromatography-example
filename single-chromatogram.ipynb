```python
# Import the `Chromatogram`
from chromatopy.core import Chromatogram
```

## Load a chromatogram from a `csv` file

```python
data_file = "data/csv/chromatogram.csv"

chrom = Chromatogram.from_csv(path=data_file, time_unit="min")
```

## Fit peaks

```python
chrom.fit()
```

# Visualize the chromatogram and the fitted peaks

```python
chrom.visualize()
```

```python
# print second peak)
print(chrom.peaks[1])
```
